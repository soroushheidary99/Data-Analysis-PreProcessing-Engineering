```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('datasets/iris.txt', names=['Sepal_lengh', 'Sepal_width', 'Petal_lengh', 'Petal_width', 'Class'])
df
```

# Descritization

```python
sl_range = [4.30, 5.55, 6.15, 7.90]
sl_labels = ['sl1', 'sl2', 'sl3']
df.Sepal_lengh = pd.cut(df.Sepal_lengh, sl_range, labels=sl_labels, retbins=False)

sw_range = [2.00, 2.95, 3.35, 4.40]
sw_labels = ['sw1', 'sw2', 'sw3']
df.Sepal_width = pd.cut(df.Sepal_width, sw_range, labels=sw_labels, retbins=False)

pl_range = [1.00, 2.45, 4.75, 6.90]
pl_labels = ['pl1', 'pl2', 'pl3']
df.Petal_lengh = pd.cut(df.Petal_lengh, pl_range, labels=pl_labels, retbins=False)

pw_range = [0.10, 0.80, 1.75, 2.50]
pw_labels = ['pw1', 'pw2', 'pw3']
df.Petal_width = pd.cut(df.Petal_width, pw_range, labels=pw_labels, retbins=False)

df
```

```python
for col in df.columns: 
    print('Number Of Unique Values Of ' + col + ' : ' + str(df[col].nunique()))
```

# ARM (Associative Rule Mining)

```python
df_onehot = df.copy()
cols = df_onehot.columns

for col in df_onehot.columns: 
    one_hot_temp = pd.get_dummies(df_onehot[col])
    df_onehot = pd.concat([df_onehot, one_hot_temp], axis=1)
    
df_onehot = df_onehot.drop(cols, axis=1)
df_onehot
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

df_apriori = apriori(df_onehot, min_support = 0.3, use_colnames = True, verbose = 1)
df_apriori
```

```python
df_ar = association_rules(df_apriori, metric = "confidence", min_threshold = 0.8)
df_ar
```
