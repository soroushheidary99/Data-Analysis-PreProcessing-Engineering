# Data cleaning is needed for furthur investigations
- many unique values are duplicated due to data entry inconsistance and difference in store labelings

#### we'll replace ', ' with ' - ' as some of the splitters have used ', ' and then split every row, using ' - ' splitter

```python
def clean(s): 
    for i in range(len(s)): 
        if ' ' in s[i]: 
            s[i] = list(s[i].split(' '))[-1]
    return s

products_clean = df.ProductName.str.replace(', ', ' - ').str.split(' - ').apply(lambda x: clean(x))
```

#### we need to one hot encode our Products name, so that we'll get to remove the uneccesary data

```python
from mlxtend.preprocessing import TransactionEncoder
trc = TransactionEncoder()
bs = trc.fit_transform( products_clean )
one_hot_clean = pd.DataFrame( bs, columns=trc.columns_mapping_.keys() )
one_hot_clean
```

### Numerical Columns have to be removed as they are meaningless in our context

```python
non_words = []
for i in one_hot_clean.columns:
    if len(i) <= 2:
        non_words.append(i)


non_letters = []
for i in one_hot_clean.columns:
    if not str.isalpha(i[0]):
        non_letters.append(i)


plurals = {}

for i in one_hot_clean.columns:
    if i[:-1] in one_hot_clean.columns and i[-1]=='s':
        plurals[i] = i[:-1]
        
```

#### there are a couple of plurals in our Products, such as both Pork and Porks, we'll merge the Porks with pork

```python
one_hot_clean = one_hot_clean.drop(non_words, axis=1, errors='ignore')
one_hot_clean = one_hot_clean.drop(non_letters, axis=1, errors='ignore')
for i, j in zip(plurals.keys(), plurals.values()): 
    one_hot_clean[j] = one_hot_clean[i] + one_hot_clean[j]
one_hot_clean = one_hot_clean.drop(list(plurals.keys()), axis=1)

one_hot_clean
```

#### There Are some Sizes and colors in our column set and also some synonyms which need to be merged such as (Untrimmed and Organic)


```python
colors_size = ['Large', 'Medium', 'Small', 'Double', 'Yellow', 'Red', 'Blue', 'Green', 'White']
one_hot_clean = one_hot_clean.drop(colors_size, axis=1, errors='ignore')


one_hot_clean['Organic'] = one_hot_clean['Organic'] + one_hot_clean['Untrimmed'] 
one_hot_clean = one_hot_clean.drop('Untrimmed', axis=1, errors='ignore')
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

df_apriori_clean = apriori(one_hot_clean, min_support = 0.003, use_colnames = True, verbose = 1)
df_apriori_clean

```

```python
df_ar_clean = association_rules(df_apriori_clean, metric = "confidence", min_threshold = 0.4)
df_ar_clean
```
